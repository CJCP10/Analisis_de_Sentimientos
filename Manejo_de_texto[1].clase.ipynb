### Funciones relevantes

```python
def clean_text(text, pattern="[^a-zA-Z0-9 ]"):
    cleaned_text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore')
    cleaned_text = re.sub(pattern, " ", cleaned_text.decode("utf-8"), flags=re.UNICODE)
    cleaned_text = u' '.join(cleaned_text.lower().split())
    return cleaned_text 
```

clean_text: Recibe dos argumentos: text y pattern. La función se encarga de limpiar y normalizar un texto para que pueda ser utilizado en un análisis posterior.

unicodedata.normalize: Normaliza el texto al formato NFD (Forma de Normalización de Combinación Canónica Descompuesta) y luego lo codifica en ASCII ignorando cualquier carácter que no pueda ser representado en ASCII.

re.sub: Utiliza expresiones regulares para reemplazar cualquier carácter que no sea una letra, un número o un espacio con un espacio en blanco. En otras palabras, elimina cualquier carácter que no sea alfanumérico o un espacio en blanco.

u ''.join(cleaned_text.lower().split()): Convierte el texto resultante en minúsculas y elimina cualquier espacio en blanco adicional.

#Conclusión: La función devuelve el texto limpio y normalizado. Este tipo de limpieza es comúnmente utilizado en tareas de procesamiento de lenguaje natural (NLP) para eliminar ruido y simplificar el texto antes de ser analizado.

```python
df["len_text_raw"] = df["CONTENT"].str.len()
df["n_dots"] = df["CONTENT"].str.split(".").str.len() - 1
df["n_artist"] = df.apply(lambda x:re.findall(x["artista"].lower(), x["CONTENT"].lower()), axis = 1).str.len()
df["n_urls"] = df.apply(lambda x:re.findall("http://", x["CONTENT"]) + re.findall("https://", x["CONTENT"]), axis = 1).str.len()
df["n_lower_ratio_len"] = df["CONTENT"].map(lambda x: sum(map(str.islower, x))) / df["len_text_raw"]
df["n_upper_ratio_len"] = df["CONTENT"].map(lambda x: sum(map(str.isupper, x))) / df["len_text_raw"]
df["n_words"] = df["CONTENT"].str.split(" ").str.len()
df["n_letters"] = df["CONTENT"].map(lambda x: sum(map(str.isalpha, x)))
df["n_lower_ratio_letters"] = df["CONTENT"].map(lambda x:sum(map(str.islower, x))) / df["n_letters"]
df["n_upper_ratio_letters"] = df["CONTENT"].map(lambda x:sum(map(str.isupper, x))) / df["n_letters"]
df["n_emojis"] = df["CONTENT"].map(emoji.emoji_count)
df["n_phone"] = df["CONTENT"].map(lambda x:re.findall(r'\+?(\d{3}[-\.\s]??\d{3}[-\.\s]??\d{4}|\(\d{3}\)\s*\d{3}[-\.\s]??\d{4}|\d{3}[-\.\s]??\d{4})', x)).str.len()
df["n_email"] = df["CONTENT"].map(lambda x: re.findall(r'[a-z0-9\._-]+@[a-z0-9\._-]+', x.lower())).str.len()
```
